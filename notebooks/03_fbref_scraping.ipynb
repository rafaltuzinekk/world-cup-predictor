```python
import pandas as pd
from pathlib import Path

# 1. Flexible Paths Configuration
BASE_DIR = Path.cwd().parent
if not (BASE_DIR / "data" / "processed").exists():
    BASE_DIR = Path.cwd()
OUTPUT_STATS_PATH = BASE_DIR / "data" / "processed" / "fbref_group_stats.csv"

print("[INFO] Loading data directly from the system clipboard...")

try:
    # 2. Pandas Magic: Read the copied HTML table (Ctrl+C)
    df_squads = pd.read_clipboard()
    
    # Identify the first column (containing country names)
    col_name = df_squads.columns[0]
    
    # 3. Clean up junk rows (mid-table headers or summary rows)
    df_squads = df_squads[df_squads[col_name] != "Squad"]
    df_squads = df_squads.dropna(subset=[col_name])
    
    # 4. Clean country names (remove prefixes like "ar ", "eng ")
    df_squads[col_name] = df_squads[col_name].astype(str).apply(lambda x: x.split(' ', 1)[-1])
    
    # Standardize the column name for merging with our ELO engine
    df_squads = df_squads.rename(columns={col_name: 'team'})
    
    # 5. Save to CSV
    df_squads.to_csv(OUTPUT_STATS_PATH, index=False)
    
    print(f"[SUCCESS] Bypassed anti-bot protection! Saved clean stats for {len(df_squads)} teams.")
    print(f"Path: {OUTPUT_STATS_PATH}")
    print("\nData Sample (Top 5):")
    print(df_squads[['team']].head().to_string(index=False))

except Exception as e:
    print(f"[ERROR] Make sure you copied the table (Ctrl+C) from FBref before running this!")
    print(f"Details: {e}")
```
